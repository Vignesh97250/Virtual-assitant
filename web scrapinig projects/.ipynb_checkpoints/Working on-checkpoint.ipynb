```python
plt.hist(data['Age'],bins=15,color='red')
plt.title("Age Distribution")
plt.show()
```

```python
counts,eadges,bars = plt.hist(data['Age'],color='magenta')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.bar_label(bars)
plt.show()
```

```python
data['Pclass'].value_counts(ascending=True).plot(kind='bar')
plt.show()
```

```python
plt.bar(data['Pclass'],data['Age'],color=['red','green'])
plt.show()
```

```python
data['Pclass'].value_counts().plot(kind='barh')
plt.show()
```

```python
data['Embarked'].value_counts().plot(kind='bar')
plt.show()
```

```python
data['Survived'].value_counts().plot(kind='bar',color='green')
plt.xticks([0,1],['Dead','Alive'])
plt.title("Survived Distribution")
plt.show()
```

```python
ax=data['Survived'].value_counts().plot(kind='bar',color='green')
plt.xticks([0,1],['Dead','Alive'])
plt.title("Survived Distribution")
for i in ax.patches:
    ax.annotate(str(i.get_height()),
                (i.get_x()+0.15,i.get_height()+0.15))
```

```python
data['Survived'].value_counts().plot(kind='pie')
plt.show()
```

```python
labels=data['Survived'].value_counts().keys().tolist()
print(labels)
values=data['Survived'].value_counts().keys().tolist()
print(values)
```

```python
plt.pie(values,labels=['Dead','Alive'],colors=['red','cyan'],
        autopct="%.2f%%")
plt.legend()
plt.title("Survived Distribution")
plt.show()
```

```python
plt.pie(values,labels=['Dead','Alive'],colors=['red','cyan'],
        startangle=90,
        autopct="%.2f%%")
plt.legend()
plt.title("Survived Distribution")
plt.show()
```

```python
plt.pie(values,labels=['Dead','Alive'],colors=['red','cyan'],
        startangle=90,
        autopct="%.2f%%")
plt.legend()
plt.title("Survived Distribution")
plt.show()
```

```python
plt.scatter(data['Age'],data['Fare'])
plt.xlabel('AgeDistribution')
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```
